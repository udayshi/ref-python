### 014.DATETIME

```python
import datetime

d=datetime.date(2018,7,24)
print(d) #2018-07-24

today=datetime.date.today()
print(today)

print(today.weekday()) #Monday 0 Sunday 6
print(today.isoweekday()) # Monday 1 Sunday 7




```

### 014.DATETIME DATETIMEDELTA
