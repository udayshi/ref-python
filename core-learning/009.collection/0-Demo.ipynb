### 009.COLLECTION DICTONARY FILTER

```python
prices = {
   'ACME': 45.23,
   'AAPL': 612.78,
   'IBM': 205.55,
   'HPQ': 37.20,
   'FB': 10.75
}

tech_names = { 'AAPL', 'IBM', 'HPQ' }

p2 = { key:value for key,value in prices.items() if key in tech_names } #{'AAPL': 612.78, 'IBM': 205.55, 'HPQ': 37.2}
print(p2)

```

### 009.COLLECTION DICTONARY LARGEST SMALLEST

```python
import heapq

portfolio = [
   {'name': 'IBM', 'shares': 100, 'price': 91.1},
   {'name': 'AAPL', 'shares': 50, 'price': 543.22},
   {'name': 'FB', 'shares': 200, 'price': 21.09},
   {'name': 'HPQ', 'shares': 35, 'price': 31.75},
   {'name': 'YHOO', 'shares': 45, 'price': 16.35},
   {'name': 'ACME', 'shares': 75, 'price': 115.65}
]

cheap = heapq.nsmallest(3, portfolio, key=lambda s: s['price']) #Cheapest 3 item
expensive = heapq.nlargest(2, portfolio, key=lambda s: s['price'])#Expensive 2 item
```

### 009.COLLECTION DICTONARY SORT

```python
from operator import itemgetter
rows = [
    {'fname': 'Brian', 'lname': 'Jones', 'uid': 1003},
    {'fname': 'David', 'lname': 'Beazley', 'uid': 1002},
    {'fname': 'John', 'lname': 'Cleese', 'uid': 1001},
    {'fname': 'Big', 'lname': 'Jones', 'uid': 1004}
]


rows_by_fname = sorted(rows, key=itemgetter('fname'))
rows_by_lfname = sorted(rows, key=itemgetter('lname','fname'))


```

### 009.COLLECTION LIST ARRAY SORT

```python
info=['uday','bijay','sanjay']
info.sort(reverse=False)
print(info)#['bijay', 'sanjay', 'uday']

#Sort based on length
info.sort(key=lambda k:len(k))
print(info) #['uday', 'bijay', 'sanjay']
```
